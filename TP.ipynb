isnull() représente les valeurs nulles avec True et dropna() supprime ces valeurs nulles directement dans le jeu de données

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
print(df2.dtypes)
```

```python
number = df2.select_dtypes(include=['int64', 'float64']).columns
letter = df2.select_dtypes(include=['object']).columns

encoder = OneHotEncoder(sparse=False)

encoded = encoder.fit_transform(df2[letter])

enco_df = pd.DataFrame(encoded, columns = encoder.get_feature_names_out(letter))

df_enco = pd.concat([df2[number], enco_df], axis=1)

correlation_matrix = df_enco.corr()

plt.figure(figsize=(27,27))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5 , linecolor='Black')
```

Les variables les plus corrélées entre elles sont les varibales avec la couleur se rapprochant le plus du rouge (et de 1).
On peut voir visuellement le résulat.
La vriable cible fumeur est moyennement corrélé avec la taille, l'hémoglobine et le genre masculin. Respectivement à 0,4 et 0,51 mais cela reste une corrélation moyenne pour ces valeurs.

```python
df2=df_enco.dropna()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X = df2.drop('smoking', axis =1)
y = df2['smoking']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

On divise en 80%, 20% car c'est les valeurs par défaut

```python
X_train.count()
```

```python
X.count()
```

```python
from sklearn.metrics import classification_report
```

```python
clf = LogisticRegression(random_state = 0, solver = 'saga', max_iter=100).fit(X_train, y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf = LogisticRegression(random_state = 0, solver = 'lbfgs', max_iter=200).fit(X_train, y_train)
clf.score(X_test,y_test)
```

```python
clf = LogisticRegression(random_state = 58,C= 100, solver = 'liblinear', max_iter=200).fit(X_train, y_train)
clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10, 100],
    'max_iter': [100, 200, 300],
    'solver': ['liblinear', 'lbfgs']
}
```

```python
model_lr = LogisticRegression()
grid_search_lr = GridSearchCV(estimator=model_lr, param_grid=param_grid_lr, cv=5, scoring='accuracy')
```

```python
grid_search_lr.fit(X_train, y_train)
```

```python
print("Best parameters for Logistic Regression: ", grid_search_lr.best_params_)
best_model_lr = grid_search_lr.best_estimator_
y_pred_lr = best_model_lr.predict(X_test)
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_pred_lr):.4f}")
```

Pour la Logistic Regression l'accuracy est de 0.7164

```python
accuracy_gb = accuracy_score(y_test, y_pred_gb)
precision_gb = precision_score(y_test, y_pred_gb, average='weighted')
recall_gb = recall_score(y_test, y_pred_gb, average='weighted')
f1_gb = f1_score(y_test, y_pred_gb, average='weighted')
cm_gb = confusion_matrix(y_test, y_pred_gb)
false_positives_gb = cm_gb.sum(axis=0) - np.diag(cm_gb)
```
